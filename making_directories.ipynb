```python
import os
import cv2
import time

cap = cv2.VideoCapture(0)
directory = 'SignImage48x48/'

while True:
    _, frame = cap.read()
    count = {letter: len(os.listdir(os.path.join(directory, letter))) for letter in "ABCDEFGHIJKLMNOPQRSTUVWXYZ"}
    count['blank'] = len(os.listdir(os.path.join(directory, "blank")))

    cv2.rectangle(frame, (0, 40), (300, 300), (255, 255, 255), 2)
    cv2.imshow("Data", frame)
    
    roi = frame[40:300, 0:300]
    cv2.imshow("ROI", roi)

    roi = cv2.cvtColor(roi, cv2.COLOR_BGR2GRAY)
    roi = cv2.resize(roi, (48, 48))

    interrupt = cv2.waitKey(10) & 0xFF
    if chr(interrupt).lower() in "abcdefghijklmnopqrstuvwxyz.":
        label = 'blank' if interrupt == ord('.') else chr(interrupt).upper()
        save_path = os.path.join(directory, label)

        print(f"Capturing 50 images for '{label}' with 3-second intervals...")
        for i in range(210):
            _, frame = cap.read()
            roi = frame[40:300, 0:300]
            roi = cv2.cvtColor(roi, cv2.COLOR_BGR2GRAY)
            roi = cv2.resize(roi, (48, 48))

            filename = os.path.join(save_path, f"{count[label] + i}.jpg")
            cv2.imwrite(filename, roi)
            print(f"Saved: {filename}")

            time.sleep(0.01)  # Wait for 3 seconds before next capture

    if interrupt == 27:  # Press 'ESC' to exit
        break

cap.release()
cv2.destroyAllWindows()
```

```python
from keras.models import model_from_json
import cv2
import numpy as np

json_file = open("2signlanguagedetectionmodel48x48.json", "r")
model_json = json_file.read()
json_file.close()
model = model_from_json(model_json)
model.load_weights("2signlanguagedetectionmodel48x48.h5")

def extract_features(image):
    feature = np.array(image)
    feature = feature.reshape(1,48,48,1)
    return feature/255.0

cv2.namedWindow("output", cv2.WINDOW_NORMAL)
cap = cv2.VideoCapture(0)
label = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'I', 'K', 'L', 'M', 'N', 'O', 'P', 'Q', 'R', 'S', 'T', 'U', 'V', 'W', 'X', 'Y', 'Z', 'blank']
while True:
    ret,frame = cap.read()
    if not ret:
        break
    cv2.rectangle(frame,(0,40),(300,300),(0, 165, 255),1)
    cropframe=frame[40:300,0:300]
    cropframe=cv2.cvtColor(cropframe,cv2.COLOR_BGR2GRAY)
    cropframe = cv2.resize(cropframe,(48,48))
    cropframe = extract_features(cropframe)
    pred = model.predict(cropframe) 
    prediction_label = label[pred.argmax()]
    cv2.rectangle(frame, (0,0), (300, 40), (0, 165, 255), -1)
    if prediction_label == 'blank':
        cv2.putText(frame, " ", (10, 30),cv2.FONT_HERSHEY_SIMPLEX,1, (255, 255, 255),2,cv2.LINE_AA)
    else:
        accu = "{:.2f}".format(np.max(pred)*100)
        cv2.putText(frame, f'{prediction_label}  {accu}%', (10, 30),cv2.FONT_HERSHEY_SIMPLEX,1, (255, 255, 255),2,cv2.LINE_AA)
    cv2.imshow("output",frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

    
cap.release()
cv2.destroyAllWindows()
```
